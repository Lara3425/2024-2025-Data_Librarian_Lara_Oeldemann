```python
import json
import pandas as pd
data1 = json.load(open("Arbeit.json"))
data2 = json.load(open("Gesundheit.json"))
```

```python
data1.keys()
```

```python
data2.keys()
```

```python
df_search1 = pd.DataFrame(data1["items"])
df_search2 = pd.DataFrame(data2["items"])
```

```python
df_search1
```

```python
df_search1["Suchbegriff"] = "Arbeit"
df_search1
```

```python
df_search2
```

```python
df_search2["Suchbegriff"] = "Gesundheit"
df_search2
```

```python
df_both = pd.concat([df_search1, df_search2])
```

```python
df_both
```

```python
df_only_publications = df_both.loc[df_both.type == "Publication"]
```

```python
df_only_publications 
```

```python
df_only_publications = df_only_publications.drop_duplicates(subset = "label", keep = "first")
```

```python
df_only_publications
```

```python
df_only_publications.columns
```

```python
df_only_publications["pages"].unique()
```

```python
df_only_publications = df_only_publications.rename(columns={"series": "Reihe", "label": "Titel"})
df_only_publications
```

```python
df_only_publications
```

```python
df_only_publications = df_only_publications.drop(columns=['user'])
df_only_publications
```

```python
df_only_publications = df_only_publications.drop(columns=['description'])
df_only_publications
```

```python
df_only_publications.columns
```

```python
df_only_publications = df_only_publications.rename(columns={"language": "Sprache", "keywords": "Schlagwort"})
df_only_publications
```

```python
df_only_publications["year_correct"] = df_only_publications.apply(lambda row: row.year.isdigit(), axis=1)
```

```python
df_only_publications.loc[df_only_publications.year_correct == False]
```

```python
df_only_publications = df_only_publications.loc[df_only_publications.year_correct == True]
```

```python
df_only_publications
```

```python
df_only_publications.columns
```

```python
df_only_publications = df_only_publications.drop(columns=['count'])
df_only_publications
```

```python
df_only_publications = df_only_publications.drop(columns=['interHash'])
df_only_publications
```
