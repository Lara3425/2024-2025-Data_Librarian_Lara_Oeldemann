```python
import requests
import pandas as pd
import numpy as np
import pickle
```

```python
query = "Kartoffel"
url = f"https://www.bibsonomy.org/json/search/{query}?items=1000"
```

```python
res = requests.get(url)
```

```python
data = res.json()
```

```python
df_Kartoffel = pd.DataFrame(data["items"])
df_Kartoffel = df_Kartoffel[df_Kartoffel["type"] == "Publication"]
```

```python
df_Kartoffel_gefiltert = df_Kartoffel.dropna(axis=1, thresh=len(df_Kartoffel)*0.5)
```
