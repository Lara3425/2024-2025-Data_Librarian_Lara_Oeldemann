```python
#Json und Pandas importieren, anschließend die heruntergeladenen Dateien einspielen
import json
import pandas as pd
data1 = json.load(open("Arbeit.json"))
data2 = json.load(open("Gesundheit.json"))
```

```python
data1.keys()
```

```python
#Datensatz anzeigen lassen aus items
data1["items"][5]
```

```python
data2.keys()
```

```python
df_search1 = pd.DataFrame(data1["items"])
df_search2 = pd.DataFrame(data2["items"])
```

```python
#Datentypen anzeigen lassen
df_search1.info()
```

```python
#extra Spalte hinzufügen
df_search1["Suchbegriff"] = "Arbeit"
df_search1
```

```python
#hier ebenfalls eine neue Spalte mit Suchbegriff hinzufügen
df_search2["Suchbegriff"] = "Gesundheit"
df_search2
```

```python
#Spalten, die in beiden Dataframes sind:
[a for a in df_search1.columns.to_list() if a in df_search2.columns.to_list()]
```

```python
#Spalten, die nur im ersten Dataframe sind:
[a for a in df_search1.columns.to_list() if a not in df_search2.columns.to_list()]
```

```python
#Spalten, die nur im zweiten Dataframe sind:
[a for a in df_search2.columns.to_list() if a not in df_search1.columns.to_list()]
```

```python
#zwei Dateien zusammenfassen
df_Gesamtdatei = pd.concat([df_search1, df_search2])
```

```python
df_Gesamtdatei
```

```python
#nur Titel anzeigen, die Publication als Typ haben
df_Publikationen = df_Gesamtdatei.loc[df_Gesamtdatei.type == "Publication"]
```

```python
df_Publikationen
```

```python
#Duplikate rausfiltern
df_Publikationen = df_Publikationen.drop_duplicates(subset = "label", keep = "first")
```

```python
df_Publikationen
```

```python
df_Publikationen = df_Publikationen.rename(columns={"series": "Reihe", "label": "Titel", "language": "Sprache", "keywords": "Schlagwort"})
df_Publikationen
```

```python
df_Publikationen.columns
```

```python
#Spalte löschen, die NA enthalten
df_NA_columns_gelöscht = df_Publikationen.dropna(axis=1)
```

```python
#Einträge auswählen, wo in Spalte "author" ein Wert eingetragen ist
df_Publikationen = df_Publikationen.loc[df_Publikationen["author"].isnull() == False]
```

```python
df_Publikationen
```

```python
#mit unique lasse ich mir die einzelnen Werte einer Spalte anzeigen
df_Publikationen["year"].unique()
```

```python
#zeilenweise einlesen, mit is digit wird überprüft ob ein String in eine Zahl umzuwandeln ist -> True oder False Bewertung
df_Publikationen["Korrektes_Jahr"] = df_Publikationen.apply(lambda row: row.year.isdigit(), axis=1)
```

```python
df_Publikationen
```

```python
#alle mit Korrektes_Jahr "False" anzeigen
df_Publikationen.loc[df_Publikationen.Korrektes_Jahr == False]
```

```python
#Datensatz überschreiben mit Korrektes Jahr True
df_Publikationen.loc[df_Publikationen.Korrektes_Jahr == True]
```

```python
df_Publikationen.columns
```

```python
df_Publikationen["Reihe"].unique()
```

```python
#beim Feld Reihe nur alles mit Inhalt anzeigen
df_Publikationen = df_Publikationen.loc[df_Publikationen["Reihe"].isnull() == False]
```

```python
df_Publikationen
```

```python
#Spalten löschen
df_Publikationen = df_Publikationen.drop(columns=['bdsk-url-1',
 'posted-at',
 'priority',
 'citeulike-article-id',
 'citeulike-linkout-0',
 'pmid',
 'translator'])
df_Publikationen
```

```python
#Spalten löschen
df_Publikationen = df_Publikationen.drop(columns=['ean',
 'asin',
 'lastchecked',
 'organisation',
 'pdf',
 'mendeley-tags',
 'faupublication',
 'peerreviewed',
 'imtm-stichworte',
 'imtm-auswahlen',
 'imtm-bscw',
 'comment',
 'contents',
 'venue',
 'eventdate',
 'eventtitle',
 'copyright',
 'date-added'])
df_Publikationen
```

```python
df_bereinigt = df_NA_columns_gelöscht[["Titel", "Autor", "Jahr", "Datum", "Reihe", "Suchbegriff", "tags", "ID", "URL"]]
df_bereinigt
```
