```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()
df = pd.read_csv("../data/Library_Usage.csv",
                                low_memory=False, na_values="Null"
)
#durch na_values= Null soll Pandas automatisch alle mit Null gekennzeichneten Werte in nan umbenennen und so erkennen. 
```

```python
df
```

```python
df["Patron Type Definition"].unique()
```

```python
#Frage 1: 
df["Patron Type Definition"].value_counts()
```

```python
df_Adults = df[df['Patron Type Definition'] == "Adult"]
```

```python

df_Teens = df[df['Patron Type Definition'] == "Teen"]
```

```python
df_Patrons = pd.concat([df_Teens, df_Adults])
```

```python
df_Patrons
```

```python
#Visualisierung zu Frage 1

#Height und Aspect nutzen, damit man den Text der X-Achse gut lesen kann!
#.set(xlim=(schiebt das Diagramm nach rechts, damit Teen nicht abgeschnitten ist))
sns.catplot(x="Patron Type Definition",
           data=df_Patrons, kind="count", color="purple", height=4, aspect=3).set(xlim=(-0.5, 3), ylim=(0, 300000), title="Bibliotheksnutzerzahl bei den Jugendlichen und Erwachsenen", xlabel="Nutzergruppe", ylabel="Anzahl")
```

```python
#Frage 2: 
df["Home Library Definition"].unique()
```

```python
#Nutzer der Hauptbibliothek zählen
df_home = df[df['Home Library Definition'] == "Main"]
```

```python
df_home
```

```python
#Visualisierung Frage 2
#Diagramm nach Nutzern der Hauptbibliothek
sns.catplot(y="Home Library Definition", hue='Age Range', 
            data=df_home, kind='count', palette=sns.color_palette("Set3")).set(title="Nutzerzahlen der Hauptbibliothek nach Altersgruppen")
```

```python
#TEST Exkurs zum Üben: Visualisierung: Kreuztabelle zu Altersspanne und Zweigbibliothek
pd.crosstab(
    df['Home Library Definition'],
    df["Age Range"],
    margins=True
)
```

```python
#TEST Exkurs zum Üben: Visualisierung
#Diagramm nach Zweigstellen
sns.catplot(x="Home Library Definition",
           data=df, kind="count", color="purple", height=4, aspect=4).set(xlim=(0, 15), ylim=(0, 150000), title="Nutzerzahlen nach Zweigstelle")
```

```python
#TEST Exkurs zum Üben: Visualisierung Ausleihen und Verlängernungen bei Senioren und Adults
sns.relplot(y='Total Checkouts', x='Total Renewals', 
                hue='Patron Type Definition', data=df_Senior_Adult)
```
