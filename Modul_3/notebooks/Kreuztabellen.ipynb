```python
import pandas as pd

df = pd.read_csv(
    "data/Library_Usage.csv", 
    low_memory=False,
    na_values="Null"
)
pd.crosstab(
    df['Notice Preference Definition'],
    df['Age Range'],
    margins=True
)
```

```python
#normalize=1 normalisiert die Spalten, heißt die Werte in der Spalte ergeben summiert 100%
#92 % der 0-9 Jährigen wählen email
pd.crosstab(
    df['Notice Preference Definition'],
    df['Age Range'],
    margins=True, normalize=1
)
```

```python
#normalize=0 normalisiert die Zeilen, heißt die komplette Zeile ergibt 100%
#von allen mailkunden befinden sich 22% in der 25-34 Jahre Klasse
pd.crosstab(
    df['Notice Preference Definition'],
    df['Age Range'],
    margins=True, normalize=0
)
```
