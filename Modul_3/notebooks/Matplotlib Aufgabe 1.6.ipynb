```python
# import the package and give it the shorter name 'plt'
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# create some dummy data
x = range(1, 10)
```

```python
# make a simple scatter plot of the data
# hier findet man noch nähere Infos: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html
plt.plot(x, x, c="black", linestyle='--', marker='+')
```

```python
#wichtig, damit überhaupt ein Diagramm angezeigt wird!
plt.show()
```
