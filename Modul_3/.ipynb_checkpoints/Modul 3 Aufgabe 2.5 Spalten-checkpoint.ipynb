```python
import pandas as pd
import csv
```

```python
# Read the CSV file
Library_data = pd.read_csv("data/Library_Usage.csv")
```

```python
df_library = pd.DataFrame(Library_data)
```

```python
df_library.columns
```

```python
df_library
```

```python
#Spalte die eigentlich Text ist, in Zahlen umwandeln und dann direkt die Spalte so Ã¼berschreiben
df_library['Circulation Active Year'] = pd.to_numeric(
  df_library['Circulation Active Year'], errors='coerce'
)
```

```python
df_library['Circulation Active Year'] 
```

```python
df_library.info()
```

```python
#Spalte ins Datumsformat umwandeln und ebenfalls Ã¼berschreiben
df_library['Circulation Active Month'] = pd.to_datetime(
    df_library['Circulation Active Month'],
    errors='coerce',
    format="%b"
)
```

```python
df_library['Circulation Active Month']
```

```python
#Extraktion des Monats als Zahl aus der Spalte
df_library['Circulation Active Month']= df_library['Circulation Active Month'].dt.month
df_library['Circulation Active Month']
```

```python
#aktive Mitgliedschaft in Monaten ausrechnen
df_library['Membership Duration'] = (df_library['Circulation Active Year'] - df_library['Year Patron Registered'])*12 + df_library['Circulation Active Month']
```

```python
df_library
```

```python
df_library['Membership Duration'] = df_library['Membership Duration'].fillna(0)
df_library['Membership Duration']
```
